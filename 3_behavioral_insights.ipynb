```python
import kagglehub
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

# Download latest version
path = kagglehub.dataset_download("priyamchoksi/credit-card-transactions-dataset")

#load up the dataset

print(os.listdir(path))

csv_path = os.path.join(path, 'credit_card_transactions.csv')

df = pd.read_csv(csv_path)

df
```

```python
#renaming and cleaning up some columns for more clarity

df = df.rename(columns= {"first": "first_name", "last": "last_name", "category": "spending_type"})
df = df.drop("Unnamed: 0", axis=1)
df
```

```python
#filling in merch_zipcode NaN values

df["merch_zipcode"] = df["merch_zipcode"].fillna("Unknown")

df
```

```python
#converting transaction dates to months

df["trans_date_trans_time"] = pd.to_datetime(df["trans_date_trans_time"])


df["month_name"] = df["trans_date_trans_time"].dt.month_name()

df["month_num"] = df["trans_date_trans_time"].dt.month

df
```

```python
#grouping by customer id and finding total spending, number of transactions and average transaction

customer_summary = (
    df.groupby("cc_num")["amt"]
      .agg(total_spend="sum", transactions="count", avg_transaction="mean")
      .reset_index()
)
customer_summary
```

```python
customer_summary.describe()
customer_summary.head()
```

```python
#comparing total spending with transactions as a scatterplot

fig, (ax1, ax2) = plt.subplots(2,1,figsize=(20,18))

sns.scatterplot(

    data=customer_summary,
    x="transactions",
    y="total_spend",
    ax=ax1,
    alpha=0.4
)

ax1.set_title("Total Spending vs Transactions")
ax1.set_xlabel("No. of Transactions")
ax1.set_ylabel("Total Spending Amount")
ax1.ticklabel_format(style="plain", axis="y")




#comparing total spending with AVERAGE transactions as a scatterplot



sns.scatterplot(

    data=customer_summary,
    x="avg_transaction",
    y="total_spend",
    ax=ax2,
    alpha=0.4
)

ax2.set_title("Total Spending vs Average Transactions")
ax2.set_xlabel("Average Transaction")
ax2.set_ylabel("Total Spending Amount")
ax2.ticklabel_format(style="plain", axis="y")
plt.tight_layout()
plt.show()
```

## ___Per-User Spending Profiles___

The relationship between total spending and transaction count shows strong
positive corelation, suggesting that customers who spend more overall tend to
do so by making purchases more often rather than through individual large
transactions.

When comparing total spending to average transaction size, the relationship is
weaker and more dispersed. High total spending customers often have moderate average
transaction values, suggesting that frequent, smaller purchases are a more
important driver of total spending than large individual transactions.

These patterns indicate the existance of distinct spending behaviors, where
high value customers are primarily defined by purchase frequency rather than
transaction size.


```python
#defining low, mid and high spenders through percentiles of total spending

p50 = customer_summary["total_spend"].quantile(0.50) #median cutomer spending
p90 = customer_summary["total_spend"].quantile(0.90) #top 10%   

p50,p90
```

```python
def type_of_spender(total_spend):       #assigning each spender to a group
    if total_spend >= p90:
        return "High Spender. (Top 10%)"
    elif total_spend >= p50:
        return "Mid Spender. (50 - 90%)"
    else:
        return "Low Spender. (Bottom 50%)"

customer_summary["spender_group"] = customer_summary["total_spend"].apply(type_of_spender)

customer_summary.head()
```

```python
customer_summary["spender_group"].value_counts(normalize=True) #normalize shows proportion instead of a raw count
```

```python
#grouping the values from spender_group and assigning total spend one by one. sum the total spend per aggregated group (low, mid, high spender)

spend_by_group = (
    customer_summary
        .groupby("spender_group")["total_spend"]
        .sum()
        .reset_index()
)

spend_by_group
```

```python
#converting to a percentage of total spending

total_spend_all = customer_summary["total_spend"].sum()

spend_by_group["spending_pct"] = (spend_by_group["total_spend"] / total_spend_all) * 100

spend_by_group
```

```python
#plotting spending contribution by a group

fig, ax = plt.subplots(figsize=(18,10))

sns.barplot(

    data = spend_by_group,
    x="spender_group",
    y="spending_pct",
    ax=ax
)

ax.set_title("Share of Total Spending Per Group")
ax.set_xlabel("Spending Groups")
ax.set_ylabel("Share of Total Spending")
ax.ticklabel_format(style="plain", axis="y")

plt.show()
```

## ___High-Spend vs Low-Spend Consumers___

Customers were segmented into low, mid, and high spenders based on total spending
percentiles. This segmentation shows that quantitaively small to normal size proportion (<40%) of
customers contributes to a large share of total spending (>50%).

Mid spenders account for a significant percentage of overall spending despite
representing a smaller fraction of users. These customers also tend to make
substantially more transactions than other segments, reinforcing the importance
of purchase frequency in driving total spending.

This distribution of spending suggests that customer behavior is highly uneven,
with revenue driven primarily by a limited group of highly active buyers.


```python
#preparing to measure volatility as coefficent of variation (std / mean)

volatility_summary = (
    df.groupby("cc_num")["amt"]
      .agg(
          avg_transaction="mean",
          transaction_std="std"
      )
      .reset_index()
)

volatility_summary.head()
```

```python
#calculating CV. we need to calculate it because it normalizes the std. std of 50 is doesnt mean the same where avg is 20 and where average is 800

volatility_summary["spending_volatility"] = volatility_summary["transaction_std"] / volatility_summary["avg_transaction"]

volatility_summary
```

```python
#merging two tables to show spender group and their spending volatility

volatility_summary = customer_summary[["cc_num","spender_group"]].merge(
    volatility_summary,
    on="cc_num",
    how="left"
)

volatility_summary.head()
```

```python
#visually showing relationship between spender_group and their spending_volatility as a boxplot

fig, ax = plt.subplots(figsize=(8,6))

sns.boxplot(

    data= volatility_summary,
    x="spending_volatility",
    y="spender_group",
    
)
ax.set_title("Spending Volatility in Spender's Groups")
ax.set_xlabel("Spending Volatility Amount")
ax.set_ylabel("")
ax.ticklabel_format(style="plain", axis="x")
```

## ___Spending Volatility Across Customer Segments___

Spending volatility was measured using the coefficient of variation of transaction
amounts, capturing how consistent or variable customer spending behavior is relative
to their average transaction size.

The results show clear differences in spending volatility across customer segments.
Low spenders exhibit the highest variability, with a wider distribution and more
extreme outliers, indicating irregular and less predictable spending behavior.
Mid spenders also display relatively high volatility, though slightly more concentrated
than the lowest-spending group.

High spenders, by contrast, tend to have lower median volatility and a more compact
distribution. While some high spending customers still show occasional extreme
variability, the majority demonstrate more stable and consistent spending patterns.
This suggests that high value customers are generally more predictable in their
purchasing behavior, despite contributing a large share of total spending.

Overall, these findings indicate that spending consistency increases with customer
value, and that volatility may be a useful behavioral signal when distinguishing
between different types of customers.


```python
#investigating whether customers spend across many different categories or just a few

category_diversity = (
    df.groupby("cc_num")["spending_type"]
      .nunique()
      .reset_index(name="num_categories")
)

category_diversity.head()
```

```python
#merging tables so they can be plotted
category_diversity = category_diversity.merge(
customer_summary[["cc_num","spender_group"]],
on="cc_num",
how="left"
    
)

category_diversity.head()
```

```python
#visualising category diversity by spender_group as a bozplot

fig, ax = plt.subplots(figsize=(8,6))

sns.boxplot(

    data=category_diversity,
    x="num_categories",
    y="spender_group",
    ax= ax
)

ax.set_title("Category Diversity by Spenders' Groups")
ax.set_xlabel("Number of Categories a Customer Spends On")
ax.set_ylabel("")
ax.ticklabel_format(style="plain", axis="x")
plt.show()
```

## ___Category Diversity Across Spender Groups___

This boxplot examines the number of unique spending categories used by customers within each spender group.

A clear pattern emerges across the segments. High spenders (top 10%) show very low variability in category diversity, with most customers consistently transacting across nearly all available categories. This results in a highly concentrated distribution near the maximum category count.

Mid spenders (50–90%) also display relatively low category diversity, almost identical to high spenders. This suggests that most mid spenders are engaged across categories.

In contrast, low spenders (bottom 50%) exhibit the greatest variability in category usage. While low spenders engage across multiple categories, some of them concentrate their activity in only a few, leading to a wider spread in the distribution.

Overall, category diversity becomes more consistent as spending increases. Higher spending customers tend to diversify naturally over time, while lower spending customers display more selective spending behavior.
