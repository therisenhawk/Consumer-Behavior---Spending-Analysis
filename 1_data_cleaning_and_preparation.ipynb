```python
import kagglehub
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

# Download latest version
path = kagglehub.dataset_download("priyamchoksi/credit-card-transactions-dataset")

#load up the dataset

print(os.listdir(path))

csv_path = os.path.join(path, 'credit_card_transactions.csv')

df = pd.read_csv(csv_path)

df
```

```python
#checking for missing values

is_missing = df.isna().any()

is_missing.value_counts()

#making a dictionary to store missing value counts
columns_with_missing_values = {}

for column in df.columns:
    missing_count = df[column].isna().sum()
    
    if missing_count > 0:
        columns_with_missing_values[column] = missing_count

columns_with_missing_values
```

```python
#renaming and cleaning up some columns for more clarity

df = df.rename(columns= {"first": "first_name", "last": "last_name", "category": "spending_type"})
df = df.drop("Unnamed: 0", axis=1)
df
```

```python
#filling in merch_zipcode NaN values

df["merch_zipcode"] = df["merch_zipcode"].fillna("Unknown")

df
```

## Dataset Overview

This dataset contains credit card transaction records showing individual purchases.

Each row corresponds to a single transaction and includes information about the transaction
amount, time, spending category, merchant location, and customer details.

The goal of this analysis is to understand consumer spending behavior by examining patterns
across categories, time, and individual users.


## Key Columns

- **Transaction Amount**:  
  The `amt` column represents the dollar amount of each credit card transaction.

- **Transaction Date and Time**:  
  The `trans_date_trans_time` column records when each transaction occurred and includes
  both date and time information.

- **Spending Category**:  
  The `spending_type` column describes the type of purchase (e.g., grocery, gas, entertainment)
  and will be used for category-level spending analysis.

- **Customer Identifier**:  
  The `cc_num` column uniquely identifies individual customers, allowing transactions
  to be grouped and analyzed at the user level.


## Data Quality and Cleaning Decisions

The dataset was inspected for duplicate records and missing values.
No duplicate transactions were found.

Missing values were present only in the `merch_zipcode` column.
Because zip code information is not required for spending pattern analysis,
these missing values were labeled as "Unknown" rather than removed.

All other columns were complete and required no additional cleaning at this stage.


## Analysis Scope

This project focuses on exploratory analysis of consumer spending behavior.
The analysis will examine spending trends across time, categories, and customers.

This project does not attempt to perform fraud detection, predictive modeling,
or credit risk assessment. Those analyses would require additional labeling
and domain-specific assumptions beyond the scope of this project.
